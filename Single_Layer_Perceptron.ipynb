```python
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/drive/MyDrive/CSE-366 (AI)/Dataset/Bank_Personal_Loan_Modelling.csv')
train_data, test_data = train_test_split(df, test_size=0.15, random_state=42)
```

```python

eta = 0.05  # Learning rate



# Training set
train_features = train_data[['Income', 'CCAvg', 'Mortgage', 'Education', 'Securities Account', 'CD Account']].values
train_labels = train_data['Personal Loan'].values

# Test set
test_features = test_data[['Income', 'CCAvg', 'Mortgage', 'Education', 'Securities Account', 'CD Account']].values
test_labels = test_data['Personal Loan'].values

# Define the necessary functions
def f(X, W):
  return W[0] + np.dot(W[1:], X)

def A(x):
  if x > 0:
    return 1
  else:
    return 0

def Y(X, W):
  return A(f(X, W))

# Define the MSE function
def MSE(Pred, T):
  Pred = np.array(Pred)
  T = np.array(T)
  err = np.mean((Pred - T) ** 2)
  return err



# Initialize weights
num_features = len(train_features[0])
w = np.zeros(num_features + 1)  # Convert to NumPy array

# Define the backpropagation function
def back_prop(TD, w, grad=True):
  Pred = []
  Ts = []
  for batch in TD:
    delw = np.zeros(len(w))  # Convert to NumPy array
    for tt in batch:
      X, T = tt
      P = Y(X, w)
      Pred.append(P)
      Ts.append(T)
      delw[0] += eta * (P - T) * 1
      delw[1:] += eta * (P - T) * X
    if grad:
      w -= delw
  return MSE(Pred, Ts)

# Define the training function
def train(TD, w):
  mse = back_prop(TD, w, grad=False)
  epoch = 0
  print("Epoch {}: Err={} and weight = {}".format(epoch, mse, w))
  while mse != 0:
    epoch += 1
    mse = back_prop(TD, w, grad=True)
    print("Epoch {}: Err={} and weight = {}".format(epoch, mse, w))

# Train the model
train_dataset = []
for i in range(len(train_features)):
  train_dataset.append((train_features[i], train_labels[i]))


TD = [train_dataset]
train(TD, w)


```

```python
# Evaluate the accuracy on the test set
correct = 0
for X, T in test_dataset:
  if Y(X, w) == T:
    correct += 1

accuracy = correct / len(test_dataset)
print("Accuracy: {:.2f}%".format(accuracy * 100))

# Print the final weights
print("Final weights:", w)
```
