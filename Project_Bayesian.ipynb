```python
def multivariate_normal (x, mean, cov):
    n = len(mean)   
    det = np.linalg.det(cov)
    norm_const = 1.0 / (np.power((2*np.pi),float(n)/2) * np.power(det,1.0/2))
    x_mu = np.matrix(x - mean)
    inv = np.linalg.inv(cov)
    result = np.power(np.e, -0.5 * (x_mu * inv * x_mu.T))
    return norm_const * result
```

```python
def Bayesian (education, securities_account, cd_account,income,cc_avg,mortagage):
  
  prob_1= DS_Prob_1(education, securities_account, cd_account,income,cc_avg,mortagage);
  # print ('Probability of taking loan',prob_1,'\n')

  prob_0=DS_Prob_0(education, securities_account, cd_account,income,cc_avg,mortagage);
  # print ('Probability of not taking loan',prob_0,'\n')

  if(prob_1>prob_0):
      # print ('Class One: Personal loan will be taken\n\nProbability of error',prob_0)
      return 1
  else:
      # print ('Class Two: Personal loan will not be taken\n\nProbability of error',prob_1)
      return 0
```

```python
def prediction_accuracy(test_df):
    accuracy = 0  # Initialize accuracy variable

    for index, row in test_df.iterrows():
        # Extract the required features from the row
        Income = row['Income']
        CCAvg = row['CCAvg']
        Mortgage = row['Mortgage']
        Education = row['Education']
        Securities_Account = row['Securities Account']
        CD_Account = row['CD Account']
        Personal_Loan = row['Personal Loan']

        # Call Naive_Bayesian function to get the prediction
        prediction = Bayesian(Education, Securities_Account, CD_Account,Income, CCAvg, Mortgage)

        if prediction== Personal_Loan:  # Compare prediction with actual value
            accuracy += 1

    return accuracy
```
